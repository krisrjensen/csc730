```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# Importing the dataset, SE330_193749481_2024_01_21.txt
dataset = pd.read_csv('SE330_193749481_2024_01_21.txt', delimiter = '\t')

#print first 10 rows
print(dataset.head(10))





```

```python
#print all field names
print(dataset.columns)

#trim all spaces and replace % with percent
dataset.columns = dataset.columns.str.strip().str.replace(' ', '_').str.replace('%', 'percent')
print(dataset.columns)
```

```python
#plot the following fields:
#NGR Current (\%FS)
#NGR Voltage (\%Setting) 	
#Calculated NGR Current (A)	
#Calculated NGR Voltage (V)	
#Nominal NGR (ohms)	
#R Change from Calibration (ohms)	
#NGR Resistance (ohms)	
#NGR Resistance (\% Nominal)	
#NGR Polarity         	
#NGR Short Mode       	
#NGR Open Mode        	
#NGR Sensitive Short Trip Level (\% Nominal)	
#NGR Sensitive Open Trip Level  (\% Nominal)	
#RF Trip Time (s)     	
#Standard RF Trip Level (ohms)	
#Pulse Enable         	
#Pulse Time (s)       	
#Vn Trip Level (V)    	
#Sensing Resistor S5 (K) 	
#GF Time (s)          	
#GF Trip Level (\%)

field_list = ['NGR_Current_(percentFS)', 'NGR_Voltage_(percentSetting)', 'Calculated_NGR_Current_(A)', 'Calculated_NGR_Voltage_(V)', 
              'Nominal_NGR_(ohms)', 'R_Change_from_Calibration_(ohms)', 'NGR_Resistance_(ohms)', 'NGR_Resistance_(percent_Nominal)',
              'NGR_Polarity', 'NGR_Short_Mode', 'NGR_Open_Mode', 'NGR_Sensitive_Short_Trip_Level_(percent_Nominal)', 'NGR_Sensitive_Open_Trip_Level_(percent_Nominal)',
                'RF_Trip_Time_(s)', 'Standard_RF_Trip_Level_(ohms)', 'Pulse_Enable', 'Pulse_Time_(s)', 'Vn_Trip_Level_(V)', 'Sensing_Resistor_S5_(K)', 'GF_Time_(s)', 'GF_Trip_Level_(percent)']

for field in field_list:
    plt.plot(dataset[field])
    plt.title(field)
    plt.show()
    


```
